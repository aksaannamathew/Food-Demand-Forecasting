## Data splitting into train and test 

```python
x=fooddemand.drop(['num_orders'],axis=1)
y=fooddemand.num_orders.values
y
```

```python
yk=np.array(y).reshape((-1, 1))
yk
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrainindp,Xtestindp,Ydeptrain,Ydeptest=train_test_split(x,yk,test_size=0.30,random_state=42)
```

```python
print(Xtrainindp.shape,Xtestindp.shape) 
print(Ydeptrain.shape,Ydeptest.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
standard=StandardScaler()
```

```python
Xtrainindp[['week','checkout_price','base_price','city_code','region_code','op_area']] = standard.fit_transform(Xtrainindp[['week','checkout_price','base_price','city_code','region_code','op_area']])
Xtrainindp
```

```python
Xtestindp[['week','checkout_price','base_price','city_code','region_code','op_area']] = standard.transform(Xtestindp[[
    'week','checkout_price','base_price','city_code','region_code','op_area']])
Xtestindp
```

```python
ydeptrain=standard.fit_transform(Ydeptrain)
ydeptrain
```

## Model Building on Linear Regression

```python
from sklearn.linear_model import LinearRegression

reg=LinearRegression()
```

```python
reg.fit(Xtrainindp,ydeptrain)
```

```python
test=reg.predict(Xtestindp)
test
```

```python
verse=standard.inverse_transform(test)
verse
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(verse, Ydeptest))  
print('Mean Squared Error:', metrics.mean_squared_error(Ydeptest,verse))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Ydeptest,verse)))
print('R square value:',metrics.r2_score(Ydeptest,verse))
```

## Random forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
Random_forest = RandomForestRegressor(n_estimators = 60, oob_score =True, n_jobs = -1,random_state =20,
                                      max_features = "auto", min_samples_leaf = 30)
```

```python
Random_forest.fit(Xtrainindp,np.ravel(ydeptrain,order='C'))
pred_Random_forest = Random_forest.predict(Xtestindp)
pred_Random_forest
```

```python
rfverse=standard.inverse_transform(pred_Random_forest)
rfverse
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(Ydeptest,rfverse))  
print('Mean Squared Error:', metrics.mean_squared_error(Ydeptest,rfverse))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Ydeptest,rfverse)))
print('R square value:',metrics.r2_score(Ydeptest,rfverse))
```

## Hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_grid={'bootstrap': [True, False],
 'max_depth': [10, 20, 25, None],
 'max_features': ['auto', 'sqrt'],
 'min_samples_leaf': [1, 2, 4],
 'min_samples_split': [2, 5, 10],
 'n_estimators': [100,200,300,400,500]}
```

```python
rf_random = RandomizedSearchCV(estimator = Random_forest, param_distributions = random_grid, n_iter = 30, n_jobs=-1,
                               cv = 3, verbose=2, random_state=42)
```

```python
rf_random.fit(Xtrainindp,np.ravel(ydeptrain,order='C'))
```

```python
print("R-Squared::{}".format(rf_random.best_score_))
print("Best Hyperparameters::\n{}".format(rf_random.best_params_))
```
